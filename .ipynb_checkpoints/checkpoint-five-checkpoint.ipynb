## Create Your Model

Use the two columns with the strongest correlation to create and train your model. Make sure to print out the summary and plot the column values and the line produced by the model.

```python
# Creating X and y
#i chose medianBaseSalary and rank 
X = df['medianBaseSalary']
y = df['rank']
#X = df['percentOfStudentsFinAid'] ----- trying different numerical values for X
#X = df['totalStudentPop']         ----- trying different numerical values for X

# Splitting the varaibles as training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, test_size = 0.3, random_state = 100)

# display train dataset
print(X_train)
print(y_train)
```

```python
# Importing Statsmodels.api library from Stamodel package
import statsmodels.api as sm

# Adding a constant to get an intercept
X_train_sm = sm.add_constant(X_train)

# Fitting the resgression line using 'OLS'
lr = sm.OLS(y_train, X_train_sm).fit()

# Printing the parameters
print("printing parameters:\n",lr.params, "\n===================================\n")

# Performing a summary to list out all the different parameters of the regression line fitted
lr.summary()
```

```python
# Visualizing the regression line
plt.scatter(X_train, y_train)
plt.plot(X_train, 862.2550 + -0.0053*X_train, 'r')
plt.show()

```

## Error Terms

Finally, plot your error terms!

```python
# Predicting y_value using train data of X
y_train_pred = lr.predict(X_train_sm)

# Creating residuals from the y_train data and predicted y_data
res = (y_train - y_train_pred)

# Plotting the histogram using the residual values
fig = plt.figure()

#sns.distplot(res, bins = 15)   
#------- `distplot` is a deprecated function and will be removed in seaborn v0.14.0. 
#HencePlease adapt your code to use either `displot` (a figure-level function with
#similar flexibility) or `histplot` (an axes-level function for histograms) 

sns.histplot(res, bins = 15, kde=True, stat="density", kde_kws=dict(cut=3))    

#---- histplot with 'kde=True' gives same result works 
#sns.displot(res, bins = 15, kde=True)      #---- displot with 'kde=True' gives same result works 
#sns.displot(res, bins = 15, kde=True, stat="density", kde_kws=dict(cut=3)) 

plt.title('Error Terms', fontsize = 15)
plt.xlabel('y_train - y_train_pred', fontsize = 15)
plt.ylabel('Density', fontsize = 15)
plt.show()


```

```python
sns.histplot(res, bins = 15, kde=True, kde_kws=dict(cut=3))    
plt.title('Error Terms', fontsize = 15)
plt.xlabel('y_train - y_train_pred', fontsize = 15)
plt.show()
```
